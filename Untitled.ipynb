## Handle Column 'Year'

### remove car-name: 'name'

```python
final_dataset = df[['year', 'selling_price', 'km_driven', 'fuel', 'seller_type',
       'transmission', 'owner', 'mileage', 'engine', 'max_power',
       'seats']]
```

```python
final_dataset.head()
```

```python
final_dataset['Current_Year']= 2020
```

```python
final_dataset.head()
```

```python
final_dataset['no_Year']= final_dataset['Current_Year']-final_dataset['year']
```

```python
final_dataset.head()
```

```python
final_dataset.drop(['year'], axis=1,inplace=True)
```

```python
final_dataset.head()
```

```python
final_dataset.drop(['Current_Year'], axis=1,inplace=True)
```

```python
final_dataset['mileage'] = final_dataset['mileage'].str.replace(" kmpl", '')
final_dataset['engine'] = final_dataset['engine'].str.replace(" CC", '')
final_dataset['max_power'] = final_dataset['max_power'].str.replace(" bhp", '')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
#Hyperparameters:
## Randomized Search CV

# Number of trees in random forest
import numpy as np
n_estimators=[int(x) for x in np.linspace(start =100, stop = 1200, num=12)]

# Number of features to consider at every split
max_features=['auto', 'sqrt']

# Maximum number if Levels in a tree
max_depth = [int(x) for x in np.linspace(5,30, num=6)]

# Minimum number of samples required to split a node
min_samples_split = [2,5,10,15,100]

# Minimum number of samples required at each leaf node
min_samples_leaf=[1,2,3,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# create the random grid

random_grid={'n_estimators': n_estimators,
             'max_features': max_features,
             'max_depth':max_depth,
             'min_samples_split': min_samples_split,
             'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf=RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator=rf, param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter =10, cv= 5, verbose =2, random_state=42, n_jobs=1)
```

```python
rf_random.fit(X_train,Y_train)
```

```python
predictions = rf_random.predict(X_test)
```
